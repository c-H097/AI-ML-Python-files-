8. **neural network layer and vectorisation**

The code below generates some data randomly, we'll treat this as our feature data (X), neuron weights (W), and neuron biases (b).

```python
# Simulate a batch for a dense layer
n_samples = 50000
d_in = 512   # input features
d_out = 512  # output units

X = np.random.randn(n_samples, d_in).astype(np.float32)
W = np.random.randn(d_in, d_out).astype(np.float32)
b = np.random.randn(d_out).astype(np.float32)
```

The code below defines two functions for an artificial neuron, one using for loops and one using numpys vectorized approach.  Comment the code to understand what's going on, then run the code and comment on the results.

```python
import time

def neuron_slow(X, W, b):
    n_samples = X.shape[0]
    out = np.empty((n_samples, W.shape[1]), dtype=np.float32)
    for i in range(n_samples):
        out[i] = X[i].dot(W) + b
    return out

def neuron_fast(X, W, b):
    return X @ W + b  # or np.matmul(X, W) + b

def benchmark(fn, n_runs=5):
    t0 = time.perf_counter()
    for _ in range(n_runs):
        out = fn(X, W, b)
    t1 = time.perf_counter()
    avg = (t1 - t0) / n_runs
    print(f"{fn.__name__}: {avg:.6f} seconds")

benchmark(neuron_slow)
benchmark(neuron_fast)
```

10. **manipulating MNIST digits with NumPy**

The code below loads the MNIST dataset (hand-written digits), then restricts it to work with only a small subset of the data.

It then:
1. Reshape the data into an array images of shape (number_of_images, 28, 28).
2. Using the labels, select all images that correspond to the digit 7 within this subset.
3. Compute the average image for the digit 7 by taking the mean across the axis that indexes samples.
4. Flatten this average image back to a vector of length 784 and compute its minimum, maximum and mean pixel values.

Comment the code to explain what is going on.


```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml("mnist_784", version=1, as_frame=False)

X_mnist = mnist["data"]
y_mnist = mnist["target"].astype(int)

X_small = X_mnist[:500]
y_small = y_mnist[:500]

images = X_small.reshape(-1, 28, 28)

sevens = images[y_small == 7]

average_seven = sevens.mean(axis=0)

average_seven_flat = average_seven.flatten()

min_pixel = average_seven_flat.min()
max_pixel = average_seven_flat.max()
mean_pixel = average_seven_flat.mean()

print("images.shape:", images.shape)
print("number of sevens in subset:", sevens.shape[0])
print("average_seven_flat length:", average_seven_flat.shape[0])
print("min_pixel:", min_pixel)
print("max_pixel:", max_pixel)
print("mean_pixel:", mean_pixel)
```
