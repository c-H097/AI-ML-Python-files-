## What type of programming language is python?

Python is a **high-level**, **interpreted**, **general-purpose** programming language.  Each of these terms highlights an important aspect of how Python works and why it has become so popular.

Being **high-level** means Python abstracts away most of the machine-level details.  You don’t need to manage memory directly or think in terms of binary instructions.  Instead, you work with clear, human-readable code that closely resembles everyday language.  This makes Python easier to learn and faster to write than many low-level languages like C.

Python is also **interpreted**, which means you don’t need to compile your code into machine language before running it.  Instead, the Python interpreter reads and executes your code line by line.  This makes experimentation quick and flexible, you can test small pieces of code interactively without waiting for a full compilation step.

It is a **general-purpose** language, which means it’s not tied to any single domain.  Python is used in web development, data science, artificial intelligence, scientific computing, automation, game development, and more.  Its extensive standard library and thriving ecosystem of third-party packages make it adaptable to almost any task.

Python is often described as object-oriented, but it also supports other paradigms like functional and procedural programming.  This flexibility allows programmers to choose the style that best fits their problem.

Another defining feature of Python is its emphasis on readability and simplicity.  Its syntax is designed to be clean and consistent, often requiring fewer lines of code than other languages to accomplish the same task.

All of these qualities make Python a beginner-friendly language while still being powerful enough for experts building large-scale applications.  It’s a language that balances ease of learning with real-world capability.

## Built-in functions

Python comes with a powerful set of **built-in functions** that are always available, no imports required.  These functions provide essential tools for working with data, performing calculations, and controlling your programs.  Learning them will make your code shorter, clearer, and more effective.

* One of the most frequently used functions is **print()**, which outputs information to the screen.  It’s invaluable for debugging and displaying results.

* The **len()** function returns the number of items in a sequence, such as a string, list, or dictionary. For instance, len("hello") returns 5.

* **type()** tells you the data type of a value, which is very helpful when debugging.  Alongside this, **isinstance()** lets you check whether a variable is of a particular type.

* Mathematical functions are also common: **abs()** gives the absolute value of a number, **round()** rounds floating-point numbers, and **sum()**, **min()**, and **max()** operate on collections.

* For creating ranges of values, **range()** is widely used with loops.  It generates sequences of integers without needing to manually build a list.

* The **input()** function allows you to capture text typed by the user, enabling interactive programs.

Other useful ones include sorted(), which returns a sorted version of a collection; enumerate(), which adds index tracking when looping through sequences; and zip(), which pairs up elements from multiple sequences.

These built-ins are just the tip of the iceberg, Python provides over 60 of them.  Familiarity with the most common ones is crucial because they provide elegant, optimized solutions that avoid reinventing the wheel.  Mastering them is a big step toward writing clean, Pythonic code.

## Keywords in python

In Python, **keywords** are reserved words that have special meaning in the language.  They form the basic syntax and structure of Python code. Since their roles are predefined, **you cannot use them as variable names or identifiers**.

Examples of keywords include **if**, **else**, **while**, **for**, **def**, and **class**.  Each signals a specific kind of operation or structure in the program. For instance, if and else are used for conditional branching, while for and while introduce loops.

There are also keywords for defining functions and classes: **def** begins a function definition, while **class** begins a class definition in object-oriented programming.

The keywords **True**, **False**, and None represent Boolean values and the absence of a value, respectively.  They are essential for logical expressions and program flow.

Control-flow keywords like **break**, **continue**, and **return** provide ways to influence loops and functions.  break exits a loop early, continue skips to the next iteration, and return exits a function while optionally sending back a result.

Some keywords are related to exception handling, such as **try**, **except**, **finally**, and **raise**.  These allow programs to respond gracefully when errors occur.

Others are tied to advanced concepts: **with** simplifies resource management, **yield** is used in generators, and **import** brings in external modules.

You don't need to know all of these now, but they are important to keep in mind.
Altogether, Python has just over 30 keywords, which makes the language easy to learn and remember.  They serve as the grammatical foundation of Python, enabling developers to express complex ideas with simple, readable code.

## Variable assignment

In Python, a variable is a name that refers to a value stored in memory. Assigning a variable is done with the assignment operator =. For example, x = 10 binds the name x to the value 10. Variables allow you to reuse and manipulate data throughout your program.

Python uses dynamic typing, which means you don’t have to declare a variable’s type. The type is inferred from the value. For example, x = 10 makes x an integer, but if you later write x = "hello", the same variable name now refers to a string.

You can assign multiple values at once. For instance, a = b = c = 0 assigns the same value to three variables, while x, y = 1, 2 assigns 1 to x and 2 to y. Python even allows swapping values without a temporary variable: x, y = y, x.

Variable names can include letters, digits, and underscores, but cannot start with a digit. They are case-sensitive (Name and name are different). For readability, use descriptive names like user_age instead of ua.

An important concept is the distinction between mutable and immutable objects. Immutable objects (like integers, floats, strings, and tuples) cannot be changed after they are created. If you “modify” one, Python actually creates a new object. Mutable objects (like lists, dictionaries, and sets) can be changed in place. This means if two variables point to the same list, changing it through one variable will also affect the other.

Understanding how assignment works is crucial for writing correct, predictable Python programs.

```python
# simple assignment
x = 10
name = "Alice"
print( x )
print( name )

# multiple assignment
a = b = c = 0
x, y = 1, 2
print( "x: ", x )
print( "y: ", y )

# swapping
x, y = y, x
print( "x: ", x )
print( "y: ", y )
```

## Data types

In Python, data types are the fundamental building blocks that determine how information is stored and manipulated.  Understanding them is essential because every operation in programming depends on the type of data you’re working with.  Let’s explore Python’s core data types.

**Numbers** are used for mathematical operations.  Python distinguishes between **integers** (int) which are whole numbers like 5, and **floats** which are decimal values like 3.14.

```python
# Numbers
x = 5
y = 3.14
print( "x type: ", type(x) )
print( "y type: ", type(y) )
```

**Strings** represent text. Strings are sequences of characters enclosed in quotes, such as "hello". They support powerful operations, including slicing, concatenation, and formatting.

```python
# Strings
text = "Hello, World!"
print( "type: ", type(text) )
print( "length: ", len(text) )
```

Lists are **ordered**, **mutable** collections of items. You can store mixed data types in a list, like \[1, "apple", 3.5\].  Since lists are mutable, you can change, add, or remove elements after creation.

```python
# Lists
fruits = ["apple", "banana", "cherry"]
print( "type: ", type(fruits) )
print( "length: ", len(fruits) )
```

**Dictionaries** store key–value pairs. They are **unordered** collections that allow fast lookups. A dictionary like {"name": "Alice", "age": 25} lets you retrieve values by referencing their keys.

```python
# Dictionaries
person = {"name": "Alice", "age": 25}
print( "type: ", type(person) )
print( "length: ", len(person) )
```

**Booleans** represent truth values: True and False. They are crucial for decision-making, often appearing as the results of comparisons or logical operations.

```python
# Booleans
is_sunny = True
print( "type: ", type(is_sunny) )
```

**Tuples** are ordered collections like lists, but they are **immutable**. Once created, their elements cannot be changed. Tuples are often used to represent fixed structures, like coordinates (x, y).

```python
# Tuples
coordinates = (10, 20)
print( "type: ", type(coordinates) )
```

**Sets** are **unordered** collections of **unique** items. They are useful when you need to eliminate duplicates or perform mathematical set operations such as union and intersection.

```python
# Sets
unique_numbers = {1, 2, 2, 3, 4}
print( "unique_numbers: ", unique_numbers )
print( "type: ", type(unique_numbers) )
```

## Lists

You can define a list in several ways. The most common is with square brackets:

```python
fruits = ["apple", "banana", "cherry"]
```

Once created, lists support many operations. You can access items by index (fruits\[0\] gives "apple") and use negative indices to count from the end (fruits\[-1\] gives "cherry"). Slicing lets you grab sublists: fruits\[0:2\] returns the first two items.

Because lists are mutable, you can modify them. Assignment by index replaces elements:

```python
print(fruits)
fruits[1] = "blueberry"
print(fruits)
```

You can also add elements with append() (adds to the end) or insert() (adds at a specific index). To combine lists, use extend() or the + operator.

Removing elements can be done with remove() (by value), pop() (by index, returning the item), or del. Lists also provide utility methods like sort(), reverse(), and count().

Overall, lists are the go-to data type when you need a flexible, ordered collection of items.

More examples:

```python
# Creating lists
fruits = ["apple", "banana", "cherry"]

# Access and slicing
print(fruits[0])    
print(fruits[-1])    
print(fruits[:2])   

# Modification
print(fruits) 
fruits[1] = "blueberry"
print(fruits) 
fruits.append("date")
print(fruits) 
fruits.insert(1, "pear")
print(fruits) 

# Removal
fruits.remove("apple")
print(fruits) 
last = fruits.pop()
print(fruits) 

# Multiple types in one list
list_multiple_types = [1,2.0,"3"]
print(list_multiple_types)
print(type(list_multiple_types))
print(type(list_multiple_types[-1]))

# Nested lists - lists within lists
nested_list = [ 0, [1,2], [3,[4,5]] ]
print(nested_list)
print(len(nested_list))
print(nested_list[1])
print(len(nested_list[1]))
print(nested_list[2])
print(len(nested_list[2]))
print(nested_list[2][1])
print(len(nested_list[2][1]))
```

## Dictionaries

The most common way to define a dictionary is with curly braces:

```python
person = {"name": "Alice", "age": 25}
```

Dictionaries are mutable, so you can add or update key–value pairs simply by assignment:

```python
print(person)
person["city"] = "Paris"
print(person)
person["age"] = 26
print(person)
```

To remove items, use del person["age"], pop("key") to remove and return a value, or clear() to empty the whole dictionary.

More examples:

```python
# Defining
person = {"name": "Alice", "age": 25}

# Updating
print(person)
person["city"] = "Paris"
print(person)
person["age"] = 26
print(person)

# Accessing
print(person['name'])
print(person['age'])
print(person['city'])

# Removing
print(person)
del person["age"]
print(person)
person.pop("city")
print(person)
```

## Comparison operations

Comparison operators are fundamental in Python because they allow you to evaluate relationships between values. Instead of just storing or displaying information, programs often need to make decisions. To do that, we compare data. A comparison always results in a Boolean value: either True or False.

The most basic operator is equality (==), which checks whether two values are the same. For example, 5 == 5 evaluates to True, while 5 == 7 evaluates to False. It’s important to note that a single equals sign = is for assignment, while == is for comparison.

Next, we have inequality (!=), which checks if two values are different. This is simply the opposite of equality.

Python also supports the familiar greater than (>), less than (<), greater than or equal to (>=), and less than or equal to (<=) operators. These are especially useful when dealing with numbers, but they also apply to strings and other comparable data types. For instance, strings are compared alphabetically based on Unicode order, so "apple" < "banana" evaluates to True.

You can also chain comparisons for more readable conditions. For example, 1 < x < 10 checks if x is between 1 and 10, inclusive of neither boundary. This style makes Python conditions more natural than in many other programming languages.

Another key point is that comparison operators integrate seamlessly with logical operators like **and**, **or**, and **not**. This lets you build more complex conditions, such as checking multiple requirements before executing code.

By mastering comparison operators, you give your programs the ability to evaluate information and make intelligent decisions, which is the essence of programming logic.

```python
# Equality and Inequality
print(5 == 5)   # True
print(5 == 5.0) # True
print(5 != 7)   # True

# Greater / Less comparisons
print(10 > 3)   # True
print(2 <= 2)   # True

# String comparison
print("apple" < "banana")  # True

# Chained comparison
x = 7
print(1 < x < 10)  # True

# Combining with logical operators
age = 20
print(age >= 18 and age < 65)  # True
```

## Conditional statements (if, elif, else)

Programs often need to make decisions: take one action if a condition is true, and another if it’s false. In Python, this is done with conditional statements, using the keywords if, elif, and else.

The basic form starts with an if statement. Python evaluates the condition after if, and if it’s True, the indented block of code beneath it executes. If the condition is False, Python skips that block. For example:

```python
temperature = 31
if temperature > 30:
    print("It's hot outside")
```

You can add an else clause to handle the opposite case:

```python
temperature = 20
if temperature > 30:
    print("It's hot")
else:
    print("It's cool")
```

When there are multiple conditions to test, Python uses elif (short for “else if”). These are checked in order, and only the first True condition’s block is executed:

```python
temperature = 22
if temperature > 30:
    print("Hot")
elif temperature > 20:
    print("Warm")
else:
    print("Cold")
```

We can combine logical operators in the condition as well:

```python
# Using logical operators
age = 18
if age >= 13 and age < 20:
    print("Teenager")
```

Conditional statements aren’t limited to numbers. Any expression that evaluates to a Boolean can be used: checking equality of strings, testing membership with in, or evaluating complex logical conditions with and/or.

Indentation is critical in Python. Unlike many languages that use braces or keywords to mark code blocks, Python relies on consistent spacing. Forgetting to indent (or mixing tabs and spaces) will cause errors.

Conditionals allow your programs to be dynamic and responsive, adapting behavior to input, environment, or user choices. They are the foundation of decision-making in code, turning static scripts into interactive logic-driven programs.

## Loops - **for** and **while**

Loops allow us to repeat a block of code multiple times, which is essential for tasks like processing collections, performing calculations, or automating repetitive work. Python provides two primary looping constructs: **for loops** and **while loops**.

A **for loop** iterates over a sequence, such as a list, string, or range of numbers. It automatically assigns each element of the sequence to a loop variable, then executes the block of code once for each element. For example:

```python
for fruit in ["apple", "banana", "cherry"]:
    print(fruit)
```

Here, fruit takes on each list element in turn. With strings, you can loop over characters. 

With the range() function, you can repeat something a specific number of times:

```python
for i in range(0,5):
    print(i)
```

A **while loop** repeats as long as a condition remains True. Unlike a for loop, which is tied to a sequence, a while loop is driven purely by logic. For example:

```python
count = 0
while count < 5:
    print(count)
    count += 1
```

Here, the loop continues until count reaches 5.

```python
countdown = 10
while countdown >= 0:
    print(countdown)
    countdown -= 1
```

Inside loops, you can use control statements:
* **break** to exit the loop early
* **continue** to skip the rest of the current iteration

Together, for and while loops give Python the flexibility to handle both data-driven repetition and condition-driven repetition, making them fundamental to all programs.

```python
# For loop with list
for fruit in ["apple", "banana", "cherry"]:
    print(fruit)

# For loop with range
for i in range(3):
    print(i)

# Enumerate
for index, value in enumerate(["a", "b", "c"]):
    print(index, value)

# While loop
count = 0
while count < 5:
    print(count)
    count += 1

# Break and continue
for n in range(10):
    if n == 5:
        break
    if n % 2 == 0:
        continue
    print(n)  # prints odd numbers less than 5
```
