We now want to check if any two people in our set of facts are actually siblings.  Below we have a nested for loop that loops through the list `father_facts`, and checks if any two people have the same father.  If they do, it prints a statement saying that those two people are siblings.

Study the code below and try to understand it.

```python
# loop over facts
for f in father_facts:
    # get the child and father from the fact using mutli-assignment 
    # note:  the underscore means we ignore the parameter in that position
    _, f_child, f_father = f 
    # now we have another loop
    # we need to use a different iterative variable, can't use 'f' again
    for g in father_facts:
        # get the child and father names again
        _, g_child, g_father = g
        # check if f_child and g_child are siblings+
        if f_father == g_father and f_child != g_child:
            # print
            print( f_child + " and  " + g_child + " are siblings" )
```

It prints twice because it finds that Leia is Luke's sibling, and Luke is Leia's sibling.

We can fix this by tracking who we've already found.  We will use lists, and order the names alphabetically so that we don't get duplicates.  Like this:

```python
siblings = []
# loop over facts
for f in father_facts:
    # get the child and father from the fact using mutli-assignment 
    # note:  the underscore means we ignore the parameter in that position
    _, f_child, f_father = f 
    # now we have another loop
    # we need to use a different iterative variable, can't use 'f' again
    for g in father_facts:
        # get the child and father names again
        _, g_child, g_father = g
        # put siblings into a list
        sl = [f_child, g_child]
        # sort the list
        sl.sort()
        # check if f_child and g_child are siblings
        if f_father == g_father and f_child != g_child and sl not in siblings:
            # add the two children names to the list 'siblings'
            siblings.append( sl )
            # print
            print( sl[0] + " and " + sl[1] + " are siblings" )
```

**Q2.3** -  Now we want a function that does this automatically for us, this is the **reasoning** part of an expert system.  Below is a function called `sibling_rule` that takes as an argument the set of facts, and returns a new list of facts that has been updated with the new rule `("siblings", {"child_1","child_2"})`, if they have the same father.

Add comments to this code to explain what is going on.

```python
def siblings_rule(facts):
    
    father_facts = []
    for f in facts:
        if f[0] == 'father':
            father_facts.append( f )

    siblings = []
    for f in father_facts:
        _, f_child, f_father = f 
        for g in father_facts:
            _, g_child, g_father = g
            sl = [f_child, g_child]
            sl.sort()
            if f_father == g_father and f_child != g_child and sl not in siblings:
                siblings.append( sl )
    
    new_facts = facts.copy()
    for s in siblings:
        sf = ("siblings",s[0],s[1])
        if sf not in new_facts:
            new_facts.add( sf )
            print("derived new fact: " + str(sf))
            
    return new_facts
```

**Q2.4** -  Next we need a rule for checking who has the force.  It's very similar.  Add comments to the function below to understand what's going on.

```python
def force_rule(facts):

    force_facts = []
    for f in facts:
        if f[0] == 'force':
            force_facts.append( f[1] )

    siblings_facts = []
    for f in facts:
        if f[0] == 'siblings':
            siblings_facts.append( f )

    for f in force_facts:
        for s in siblings_facts:
            if f in s:
                if s[1] not in force_facts:
                    print("derived new fact: " + str(("force",s[1]) ))
                    force_facts.append(s[1])
                if s[2] not in force_facts:
                    print("derived new fact: " + str(("force",s[2]) ))
                    force_facts.append(s[2])

    new_facts = facts.copy()
    for f in force_facts:
        new_facts.add( ('force',f) )

    return new_facts        
```

**Q2.6** -  Now we will use a simple forward-chaining algorithm that uses this function to take all initial facts and rules to their logical conclusions.  Comment this code to explain what is going on.

```python
def forward_chain(facts, rules):
    new_facts = facts.copy()
    print( "starting facts: " + str(facts) )
    for i in range(10):
        for rule in rules:
            new_facts = rule(new_facts)
    return new_facts
```

```python
new_facts = forward_chain( facts, [siblings_rule, force_rule] )
```
