## 2) Basics: Events & Simple Probability
In a uniform sample space, the probability of an event is:

**Probability = (number of favourable outcomes) / (total outcomes).**

Example: On a fair die, there are 6 equally likely faces.

```python
# Helper function for uniform spaces
def p_uniform(event_size, space_size):
    return event_size / space_size

# Die examples (uniform):
space_size = 6
p_even = p_uniform(3, space_size)        # {2,4,6}
p_ge5  = p_uniform(2, space_size)        # {5,6}
p_six  = p_uniform(1, space_size)
p_even, p_ge5, p_six
```

### Inclusion–Exclusion on a die
For events A = 'even' and B = '≥5', their intersection is {6}. The inclusion–exclusion law says:
$$P(A\cup B)=P(A)+P(B)-P(A\cap B).$$
Let’s compute it:

```python
p_even_or_ge5 = p_even + p_ge5 - p_uniform(1, space_size)
p_even_or_ge5
```

## 3) Conditional Probability & Independence
**Conditional probability:** $P(A\mid B) = P(A\cap B)/P(B)$ (when $P(B)>0$). It means: focus on the world where B happened, and ask how often A also happens.

**Independence:** A and B are independent if $P(A\cap B)=P(A)P(B)$ (equivalently $P(A\mid B)=P(A)$).

```python
# Study vs pass example (2x2 table)
passed_study = 120
study = 150
passed_nostudy = 36
nostudy = 90
p_pass_given_study = passed_study / study
p_pass_given_nostudy = passed_nostudy / nostudy
p_overall_pass = (study/240)*p_pass_given_study + (nostudy/240)*p_pass_given_nostudy
p_pass_given_study, p_pass_given_nostudy, p_overall_pass
```

## 4) Bayes’ Rule
Bayes’ theorem updates a hypothesis probability after seeing data:
$$P(H\mid D)=\frac{P(D\mid H)P(H)}{P(D)}\quad\text{where}\quad P(D)=\sum_i P(D\mid H_i)P(H_i).$$
Think: **Prior × Likelihood → Posterior**; the **evidence** $P(D)$ ensures results sum to 1.

### 4.1 Medical Test (one positive)
Scenario: Prevalence (prior) = 0.01, Sensitivity = 0.95, Specificity = 0.95. Compute `P(Disease | Positive)` and interpret.

```python
prev = 0.01
sens = 0.95
spec = 0.95
fpr  = 1 - spec
pd_pos = sens*prev + fpr*(1 - prev)    # evidence P(+)
ppv = (sens*prev) / pd_pos             # P(Condition|+)
ppv
```

**Interpretation:** With a rare condition (1% prevalence), even a good test yields a posterior around 16%. This is the false-positive effect at low base rates.

## 5) Bayes with Dice: Fair vs Loaded
We have two hypotheses about a die:
- $H_L$ (Loaded): P(6) = 0.5, the remaining 0.5 is shared among {1,2,3,4,5}.
- $H_F$ (Fair): P(6) = 1/6.

Priors: P($H_L$)=0.2, P($H_F$)=0.8. We observe die rolls and update our belief in $H_L$.

### 5.1 One roll is a 6 — compute the posterior P(Loaded | 6)

```python
pL, pF = 0.2, 0.8
p6_L, p6_F = 0.5, 1/6
evidence = p6_L*pL + p6_F*pF
post_L_given_6 = (p6_L*pL) / evidence
post_L_given_6
```

### 5.2 Two sixes — assume independence given the hypothesis
Compute `P(Loaded | 6,6)` in one shot by squaring the likelihoods.

```python
evidence2 = (p6_L**2)*pL + (p6_F**2)*pF
post_L_given_66 = (p6_L**2 * pL) / evidence2
post_L_given_66
```

## 6) Monte Carlo Simulations (advanced and optional)

### 6.1 Dice simulation — estimate P(Loaded | roll = 6)
We randomly choose a die (loaded with probability 0.2, fair otherwise), roll once, and estimate the chance it was the loaded die given the outcome is 6.

```python
def roll_loaded(rng):
    """
    simulate one roll of a *loaded* die.
    rule: P(6) = 0.5. the remaining 0.5 probability is split uniformly over {1,2,3,4,5}.
    """
    # draw a uniform random number in [0,1).
    # if it's < 0.5, we say the outcome is 6 (this happens with probability 0.5).
    if rng.random() < 0.5:
        return 6
    # otherwise (also with probability 0.5), we choose uniformly from {1,2,3,4,5}.
    # note: rng.integers(1, 6) returns an integer in {1,2,3,4,5} because the upper bound is exclusive.
    return int(rng.integers(1, 6))

def roll_fair(rng):
    """
    simulate one roll of a *fair* die: each face 1..6 has probability 1/6.
    """
    # rng.integers(1, 7) returns an integer in {1,2,3,4,5,6}.
    return int(rng.integers(1, 7))

def simulate_once(rng, p_loaded=0.2):
    """
    one trial of the experiment:
      1) first choose which die we picked (loaded vs fair) using the prior P(loaded) = p_loaded.
      2) then roll the chosen die once.
      3) return both the hidden truth (is_loaded) and the observed roll.
    """
    # is_loaded is True with probability p_loaded, False otherwise.
    is_loaded = rng.random() < p_loaded

    # conditional on which die was chosen, generate one observed roll.
    if is_loaded:
        roll = roll_loaded(rng)
    else:
        roll = roll_fair(rng)

    return is_loaded, roll

# number of Monte Carlo trials (bigger N -> lower Monte Carlo noise).
N = 200000

# pre-allocate arrays to store the hidden truth (is the die loaded?) and the observed roll each trial.
is_loaded_arr = np.empty(N, dtype=bool)
roll_arr = np.empty(N, dtype=int)

# run N independent trials of the experiment.
for i in range(N):
    is_loaded_arr[i], roll_arr[i] = simulate_once(rng)

# build a boolean mask for the trials where we *observed* a 6.
mask6 = (roll_arr == 6)

# among those trials with roll == 6, compute the fraction that actually came from the *loaded* die.
# in NumPy, .mean() on a boolean array gives the proportion of True values.
est = is_loaded_arr[mask6].mean()

# 'est' is our Monte Carlo estimate of P(Loaded | roll = 6).
est
```

```python
def simulate_two_rolls(rng, p_loaded=0.2):
    """
    One trial of the *two-roll* experiment.
      1) Pick which die (loaded vs fair) with P(loaded) = p_loaded.
      2) Roll it twice.
      3) Return (is_loaded, roll1, roll2).
    """
    # choose which die was picked
    is_loaded = rng.random() < p_loaded

    # roll twice with the same die
    if is_loaded:
        roll1 = roll_loaded(rng)
        roll2 = roll_loaded(rng)
    else:
        roll1 = roll_fair(rng)
        roll2 = roll_fair(rng)

    return is_loaded, roll1, roll2


# number of trials
N = 200000
is_loaded_arr = np.empty(N, dtype=bool)
roll1_arr = np.empty(N, dtype=int)
roll2_arr = np.empty(N, dtype=int)

# simulate N independent experiments
for i in range(N):
    is_loaded_arr[i], roll1_arr[i], roll2_arr[i] = simulate_two_rolls(rng)

# build a mask for the *data event* — both rolls were 6
mask_both6 = (roll1_arr == 6) & (roll2_arr == 6)

# among those with (6,6), estimate the fraction that truly came from the loaded die
est_two = is_loaded_arr[mask_both6].mean()

est_two
```

### 6.2 Medical test simulation — estimate PPV
We simulate many patients with prevalence 1%. Each gets a test with sensitivity 95% and specificity 95%. Among those who test positive, we measure the fraction who actually have the disease (PPV).

Comment the below code to explain what is going on, like the code above.

```python
def simulate_test(rng, prev=0.01, sens=0.95, spec=0.95):
    has = rng.random() < prev
    if has:
        pos = rng.random() < sens
    else:
        pos = rng.random() < (1 - spec)
    return has, pos

N = 300000
has_arr = np.empty(N, dtype=bool)
pos_arr = np.empty(N, dtype=bool)
for i in range(N):
    has_arr[i], pos_arr[i] = simulate_test(rng, prev=0.01, sens=0.95, spec=0.95)
ppv_sim = has_arr[pos_arr].mean()
ppv, ppv_sim
```

### 6.3 Monty Hall simulation — switch vs stay and Bayes check
Rules: The host knows where the car is, always opens a goat door you didn’t pick, and if he has a choice he picks uniformly at random. We estimate the win rates for **stay** vs **switch** strategies and empirically check the likelihoods used in a Bayes solution.

Comment the code below to understand what's going on, the result should match what we learned in the lectures!

```python
def monty_trial(rng):
    doors = np.array([0,1,2])
    car   = int(rng.integers(0,3))     # car location
    pick  = 0                          # player picks door 0
    goat_doors = doors[(doors != pick) & (doors != car)]
    if goat_doors.size == 0:
        host = int(rng.choice(doors[doors != pick]))
    else:
        host = int(rng.choice(goat_doors))
    # find the only door that's not the pick and not the host's door
    switch = None
    for d in doors:
        if d != pick and d != host:
            switch = d
            break
    win_stay   = (pick == car)
    win_switch = (switch == car)
    opened = host
    return win_stay, win_switch, car, opened

N = 200000
stay_wins = 0
switch_wins = 0
opened3_count = 0
opened3_given_car1 = 0
opened3_given_car2 = 0
opened3_given_car3 = 0

for _ in range(N):
    ws, wsw, car, opened = monty_trial(rng)
    stay_wins   += ws
    switch_wins += wsw
    if opened == 2:
        opened3_count += 1
        if car == 0: opened3_given_car1 += 1
        if car == 1: opened3_given_car2 += 1
        if car == 2: opened3_given_car3 += 1

stay_rate   = stay_wins   / N
switch_rate = switch_wins / N
p_D = opened3_count / N
p_D_given_H1 = opened3_given_car1 / (N/3)
p_D_given_H2 = opened3_given_car2 / (N/3)
p_D_given_H3 = opened3_given_car3 / (N/3)

stay_rate, switch_rate, p_D, p_D_given_H1, p_D_given_H2, p_D_given_H3
```

**Interpretation:** The `switch` strategy wins about two-thirds of the time. Empirical likelihoods for data `D = host opens door 3` are approximately `P(D)=0.5`, `P(D|car=1)=0.5`, `P(D|car=2)=1.0`, `P(D|car=3)=0.0`, matching the Bayes analysis (posterior 2/3 on the other unopened door).

### 6.4 Monty-Hall experiment with more doors!

Let's say we have more doors in the game.  And instead of opening just one door, Monty opens all doors except one, containing a goat.  How does this change the outcome?  How does it effect the odds before or after switching?

```python
def monty_hall(T=3, N=100000, seed=None):
    """
    simulate the Monty Hall experiment for any number of doors T ≥ 3.
    
    monty always opens T−2 doors that don't contain the prize and weren't picked,
    leaving two closed doors: your original pick and one other.
    
    prints the success rates for staying vs switching.
    """
    assert T >= 3, "T must be at least 3"
    rng = np.random.default_rng(seed)

    # 1) randomly assign prize doors and initial picks
    prizes = rng.integers(0, T, size=N)   # each element = prize door index
    picks  = rng.integers(0, T, size=N)   # each element = player's first choice

    # 2) staying wins when pick == prize
    stay_wins = np.sum(picks == prizes)

    # 3) switching wins when pick != prize
    switch_wins = N - stay_wins

    # 4) compute success rates
    stay_rate = stay_wins / N
    switch_rate = switch_wins / N

    # 5) print results
    print(f"T = {T}, runs = {N}")
    print(f"Stay win rate:   {stay_rate:.3f}")
    print(f"Switch win rate: {switch_rate:.3f}")

# Example usage:
# monty_hall(T=3, N=200000)
# monty_hall(T=10, N=200000)
```

## 7) Common Pitfalls & Checks
- **Base-rate neglect:** PPV strongly depends on prevalence (the prior).
- **Confusing conditionals:** Don’t mix up P(A|B) with P(B|A).
- **Assuming independence** without justification.
- **Sanity checks:** Does the posterior move the way you expect when data favours a hypothesis?
