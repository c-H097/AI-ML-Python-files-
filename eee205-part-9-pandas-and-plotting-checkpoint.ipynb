4 - Count the number of vowels (a e i o u) in the string below.  The results should be case insensitive.

```python
s = "Abracadabra"
vowels = ["a","e","i","o","u"]
#s = s.lower()
count = 0

for char in s:
    if char in vowels:
        count += 1
        
print(count)      
```

5 - Write a function that checks if a string is a palindrome (the same backwards and forwards), if it is True then retrun True, if not then return False.  It should be case insensitive.

Hint:  use word\[::-1\]

```python
def palindrome(word):
    word = word.lower()
    if word == word[::-1]:  #reverses string
        print("palindrome")
    else:
        print("not a palindrome")
 
```

```python
palindrome("Cat")
```

6 - Generate and print all Fibonacci numbers less than or equal to 100 using a while loop.

The Fibonacci sequence is a series of numbers where each number is the sum of the two preceding numbers.
The first two numbers are \[0,1\].

```python
fn_list = [0,1]
```

```python
while True:
    fn = fn_list[-2] + fn_list[-1]
    if fn <= 100:
        fn_list.append(fn)
    else:
        break

print(fn_list)
     
    
```
