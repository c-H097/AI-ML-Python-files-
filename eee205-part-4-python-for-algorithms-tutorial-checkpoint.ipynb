# The river crossing problem

A farmer must take a wolf, a goat, and a cabbage across a river.
- The boat can carry the farmer + one passenger
- But.. if the wolf is left with the goat, it’ll eat the goat.
- And if the goat is left with the cabbage, it’ll eat the cabbage

How can the farmer get all three across the river?

**We'll solve this by implementing a Breadth-First-Search (BFS)**

### Step 1: define the problem

The first thing we need to do is to define the 'state' of the system.

The system here is just locations of the farmer, the wolf, the goat, and the cabbage.  The whole puzzle is about moving each of these players from one side of the river to the other, and the constraints constrain where certain players can be at any one time.

So we can define the state of the system with a **tuple**!

The tuple will have four entries containing the positions of the four players.  We can choose the ordering to be anything we like, let's say it is `(farmer,wolf,goat,cabbage)`.  Then if we had the state `state = (L,R,L,R)` it would mean that the farmer and the goat are on the left (L) side of the river, and the wolf and cabbage are on the right (R) side of the river.

Q1. Define the **initial state** for the problem:

```python
initial_state = ("L", "L", "L", "L")
```

The **state space** is then just all possible combinations of left (L) and right (R) in the tuple.  E.g. we could have a bunch of possible states like `(L,R,R,R)`, `(R,R,L,L)`, etc.  Infact, there are $2^4=32$ possible combinations.

However, we have **constraints** that mean some of these states aren't allowed.

```python
#def is_valid(farmer, wolf, goat, cabbage):
def is_valid(state):
    #unpack the state tuple into individual variables
    farmer, wolf, goat, cabbage = state
    
    # check if wolf would eat goat (when farmer is not present)
    if farmer != wolf and wolf == goat:
        return False
    
    #check if the goat eats the cabbage when the farmer is not present
    if farmer != goat and goat == cabbage:
        return False


    return True
state = "L", "L", "R", "R"
is_valid(state)

# returns false if the wolf and goat are on the same side WITHOUT the farmer
# the goat and cabbage are on the same side WITHOUT the farmer
# Otherwise, it returns True, meaning the state is valid.
```

### Step 2: Breadth-First-Search

In the BFS, we explore all possibile next moves at the same time.

So, if we're in some `state`, sometimes called a 'configuration', we need to be able to work out all the possible next moves.

```python
def get_next_states(state, actions):
    '''
    The possible actions are:
    - farmer goes alone
    - farmer takes wolf
    - farmer takes goat
    - farmer takes cabbage
    '''
    #current state
    farmer, wolf, goat, cabbage = state
   # state =  farmer, wolf, goat, cabbage
    # initialise a list to store the next states in
    next_states = []   # changed == to =

    # now we consider each possible action separately
    actions = [
        'farmer_goes_alone',
        'farmer_takes_wolf',
        'farmer_takes_goat',
        'farmer_takes_cabbage'
    ]

    # loop through each possible action
    for action in actions:

        # farmer going alone
        if action == 'farmer_goes_alone':
            #farmer moving to other side, updating side
            if farmer == "L":
                new_farmer = "R"
            else:
                new_farmer = "L"
            #create new state where only farmer moved
            new_state = (new_farmer, wolf, goat, cabbage)
            #checks if this state is valid
            if is_valid(new_state):
                next_states.append( new_state )

        #farmer taking the wolf
        elif action == 'farmer_takes_wolf':
            #check if wolf is on same side as farmer
            if farmer == wolf:
                #move both farmer and wolf
                if farmer == "L":
                    new_farmer = "R"
                    new_wolf = "R"
                else:
                    new_farmer = "L"
                    new_wolf = "L"
                #create new state
                new_state = (new_farmer, new_wolf, goat, cabbage)
                #check if valid
                if is_valid(new_state):
                    next_states.append( new_state )
                

        #farmer taking a goat
        elif action == 'farmer_takes_goat':
            #check if goat is on same side as farmer
            if farmer == goat:
                # Move both farmer and goat
                if farmer == "L":
                    new_farmer = "R"
                    new_goat = "R"
                else:  # Fixed: elif to else
                    new_farmer = "L"
                    new_goat = "L"
            #
            new_state = (new_farmer, wolf, new_goat, cabbage)
            #
            if is_valid(new_state):
                next_states.append( new_state )

        #
        elif action == 'farmer_takes_cabbage':
            #
            if farmer == cabbage:
                # Move both farmer and cabbage
                if farmer == "L":
                    new_farmer = "R"
                    new_cabbage = "R"
                else:
                    new_farmer = "L"
                    new_cabbage = "L"
            #
            new_state = (new_farmer, wolf, goat, new_cabbage)
            #
            if is_valid(new_state):
                next_states.append( new_state )
    
    return next_states
state = "L", "L", "L", "L"    
action = 'farmer_takes_wolf'
get_next_states(state, action)
```
